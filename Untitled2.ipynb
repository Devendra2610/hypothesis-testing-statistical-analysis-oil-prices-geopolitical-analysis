### What is Hypothesis Testing?

Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It's a formal procedure for investigating our ideas about the world using statistics. It is most often used by scientists to test specific predictions, called hypotheses, by calculating how likely it is that a pattern or relationship observed in a sample occurred by chance.

The general process involves:
1.  **Formulating Hypotheses:** Stating a null hypothesis (H₀) and an alternative hypothesis (H₁).
2.  **Setting the Significance Level (α):** Deciding on the probability threshold for rejecting the null hypothesis (commonly 0.05).
3.  **Collecting Data:** Gathering relevant data through experiments or observations.
4.  **Calculating Test Statistic:** Computing a value from the sample data that summarizes the evidence against the null hypothesis.
5.  **Making a Decision:** Comparing the p-value (or test statistic) to the significance level to decide whether to reject or fail to reject the null hypothesis.

### Null Hypothesis (H₀) vs. Alternative Hypothesis (H₁)

In hypothesis testing, we always set up two competing hypotheses:

*   **Null Hypothesis (H₀):** This is the statement that there is no effect, no difference, or no relationship between variables. It represents the status quo or a default assumption. We assume the null hypothesis is true until we have sufficient evidence to reject it. For example, H₀: The average height of men is 175 cm.

*   **Alternative Hypothesis (H₁ or Hₐ):** This is the statement that there is an effect, a difference, or a relationship between variables. It's what the researcher is trying to prove or find evidence for. It contradicts the null hypothesis. For example, H₁: The average height of men is NOT 175 cm.

### Types of Alternative Hypotheses (One-sided vs. Two-sided)

The alternative hypothesis can be directional or non-directional, leading to different types of tests:

*   **Two-sided Hypothesis (Non-directional):** This type of hypothesis suggests that there is a difference or effect, but doesn't specify the direction of the difference. The alternative hypothesis states that the parameter is *not equal to* a specific value.
    *   Example: H₀: The average test score is 75. H₁: The average test score is ≠ 75.

*   **Left-sided Hypothesis (Directional):** This type of hypothesis suggests that the true parameter is *less than* a specific value. It's used when we are specifically interested in a decrease or a value being below a certain point.
    *   Example: H₀: The average reaction time is 2 seconds. H₁: The average reaction time is < 2 seconds.

*   **Right-sided Hypothesis (Directional):** This type of hypothesis suggests that the true parameter is *greater than* a specific value. It's used when we are specifically interested in an increase or a value being above a certain point.
    *   Example: H₀: The average yield of the new fertilizer is 100 kg. H₁: The average yield of the new fertilizer is > 100 kg.

### Null Hypothesis vs. Alternative Hypothesis: Which Statistical Test to Choose?

The choice of a statistical test depends on several factors:

1.  **Type of Data:** (e.g., categorical, numerical, ordinal).
2.  **Number of Samples/Groups:** (e.g., one sample, two independent samples, paired samples, more than two samples).
3.  **Distribution of Data:** (e.g., normal distribution, skewed distribution).
4.  **Research Question:** What specifically are you trying to determine (e.g., comparing means, comparing proportions, testing for associations)?
5.  **Assumptions of the Test:** Many tests have specific assumptions that must be met (e.g., normality, homogeneity of variances).

Below, I'll explain some popular statistical tests and demonstrate their use.

### Most Popular Types of Statistical Tests in Data Science

Here are some of the most commonly used statistical tests:

#### Z-test for Population Mean

A Z-test is used to test a hypothesis about a population mean when the population standard deviation is known and the sample size is large (typically n > 30). It assumes that the data is normally distributed.

**When to use:** When you want to compare a sample mean to a known population mean (or a hypothesized population mean) and you know the population standard deviation.

**Example Scenario:** A manufacturing company claims their light bulbs have an average lifespan of 10,000 hours with a population standard deviation of 500 hours. A sample of 100 bulbs has an average lifespan of 9,800 hours. Is the company's claim accurate?

Let's assume the `oil_geopolitics_df`'s '`oil_price`' column represents some sample data, and we want to test if its mean is significantly different from a hypothesized population mean, assuming we know the population standard deviation. For demonstration, let's hypothesize the population mean oil price is $80, and the population standard deviation is known to be $10.

```python
from statsmodels.stats.weightstats import ztest
import numpy as np

# Let's use the 'brent_price' column from the oil_geopolitics_df as our sample data
# First, check if the column exists and has numerical values
if 'brent_price' in oil_geopolitics_df.columns:
    sample_data = oil_geopolitics_df['brent_price'].dropna()
    if not sample_data.empty:
        # Hypothesized population mean
        pop_mean_hypothesized = 80
        # Known population standard deviation (for demonstration, assume it's 10)
        pop_std_dev = 10

        # Let's manually calculate the Z-statistic and p-value using the known population standard deviation
        sample_mean = sample_data.mean()
        n = len(sample_data)
        std_error = pop_std_dev / np.sqrt(n)
        z_statistic = (sample_mean - pop_mean_hypothesized) / std_error

        from scipy.stats import norm
        p_value = 2 * (1 - norm.cdf(abs(z_statistic))) # Two-tailed p-value

        print(f"Sample Mean Brent Oil Price: ${sample_mean:.2f}")
        print(f"Hypothesized Population Mean: ${pop_mean_hypothesized}")
        print(f"Calculated Z-statistic: {z_statistic:.3f}")
        print(f"P-value: {p_value:.3f}")

        alpha = 0.05
        if p_value < alpha:
            print(f"At α={alpha}, we reject the null hypothesis. There is significant evidence that the true population mean Brent oil price is different from ${pop_mean_hypothesized}.")
        else:
            print(f"At α={alpha}, we fail to reject the null hypothesis. There is not enough significant evidence to say that the true population mean Brent oil price is different from ${pop_mean_hypothesized}.")
    else:
        print("The 'brent_price' column is empty after dropping NaN values.")
else:
    print("The 'brent_price' column was not found in the oil_geopolitics_df. Cannot perform Z-test demonstration.")
```

#### One-Sample t-test for Population Mean

A one-sample t-test is used to determine whether a sample mean is statistically different from a known or hypothesized population mean when the population standard deviation is *unknown*. It is appropriate for small sample sizes (n < 30) or when the population standard deviation cannot be assumed.

**When to use:** To compare a sample mean to a hypothesized population mean when the population standard deviation is unknown.

**Example Scenario:** A new teaching method is introduced, and a sample of 25 students using this method achieved an average score of 85. The historical average score for the old method was 80. Is the new method significantly better?

Let's continue using the `oil_geopolitics_df['oil_price']` as our sample and test if its mean is significantly different from a hypothesized value, assuming the population standard deviation is *unknown*.

```python
from scipy import stats

if 'brent_price' in oil_geopolitics_df.columns:
    sample_data = oil_geopolitics_df['brent_price'].dropna()
    if not sample_data.empty:
        # Hypothesized population mean
        pop_mean_hypothesized = 80 # Let's keep the same hypothesized mean as for Z-test

        # Perform one-sample t-test
        t_statistic, p_value = stats.ttest_1samp(sample_data, pop_mean_hypothesized)

        print(f"Sample Mean Brent Oil Price: ${sample_data.mean():.2f}")
        print(f"Hypothesized Population Mean: ${pop_mean_hypothesized}")
        print(f"Calculated t-statistic: {t_statistic:.3f}")
        print(f"P-value: {p_value:.3f}")

        alpha = 0.05
        if p_value < alpha:
            print(f"At α={alpha}, we reject the null hypothesis. There is significant evidence that the true population mean Brent oil price is different from ${pop_mean_hypothesized}.")
        else:
            print(f"At α={alpha}, we fail to reject the null hypothesis. There is not enough significant evidence to say that the true population mean Brent oil price is different from ${pop_mean_hypothesized}.")
    else:
        print("The 'brent_price' column is empty after dropping NaN values.")
else:
    print("The 'brent_price' column was not found in the oil_geopolitics_df. Cannot perform One-Sample t-test demonstration.")
```

#### Paired t-Test

A paired t-test (also called a dependent t-test) is used to compare the means of two related groups or samples. This typically occurs when the same subjects are measured twice (e.g., 'before' and 'after' an intervention) or when subjects are matched in pairs.

**When to use:** To compare the means of two related measurements from the same individuals or matched pairs.

**Example Scenario:** A researcher wants to assess the effectiveness of a new diet plan. They measure the weight of 15 individuals before starting the diet and after 3 months on the diet. Is there a significant difference in weight after the diet?

For demonstration, since our current datasets don't have clear 'before' and 'after' measurements, we'll create some synthetic paired data.

```python
from scipy import stats
import numpy as np

# Generate synthetic 'before' and 'after' data for demonstration
np.random.seed(42)
before_treatment = np.random.normal(loc=70, scale=5, size=20) # e.g., weight before diet
after_treatment = before_treatment - np.random.normal(loc=2, scale=1, size=20) # e.g., weight after diet, with some reduction

# Perform paired t-test
t_statistic, p_value = stats.ttest_rel(before_treatment, after_treatment)

print(f"Mean 'Before' Treatment: {np.mean(before_treatment):.2f}")
print(f"Mean 'After' Treatment: {np.mean(after_treatment):.2f}")
print(f"Calculated t-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print(f"At α={alpha}, we reject the null hypothesis. There is a significant difference between 'before' and 'after' treatment means.")
else:
    print(f"At α={alpha}, we fail to reject the null hypothesis. There is no significant difference between 'before' and 'after' treatment means.")
```

#### Two-Sample t-Test (Independent Samples t-Test)

A two-sample t-test (also known as independent samples t-test) is used to compare the means of two independent groups to determine if there is a statistically significant difference between them.

**When to use:** To compare the means of two separate, unrelated groups.

**Assumptions:** Independent observations, normal distribution, and (often) equal variances (though there are versions for unequal variances).

**Example Scenario:** A study compares the effectiveness of two different fertilizers on crop yield. One group of crops receives fertilizer A, and another independent group receives fertilizer B. Is there a significant difference in crop yield between the two fertilizers?

For demonstration, let's imagine we can split the `oil_price` data into two groups based on a geopolitical event, or we can simply simulate two independent groups of oil prices.

```python
from scipy import stats
import numpy as np

# For a more realistic example with the provided data, we could try to split 'oil_price' by some geopolitical event.
# However, the 'geopolitical_events_timeline.csv' has event descriptions, not a simple binary 'event/no-event' flag
# that can be directly mapped to the oil_geopolitics_df time series. Also, oil_geopolitics_df does not have a direct
# 'event' column. So, let's create two synthetic independent groups for clear demonstration.

np.random.seed(42)
group_a_oil_prices = np.random.normal(loc=75, scale=8, size=50) # e.g., oil prices during Period A
group_b_oil_prices = np.random.normal(loc=80, scale=9, size=60) # e.g., oil prices during Period B

# Perform independent two-sample t-test
# We use equal_var=False for Welch's t-test, which does not assume equal population variance
t_statistic, p_value = stats.ttest_ind(group_a_oil_prices, group_b_oil_prices, equal_var=False)

print(f"Mean Oil Price Group A: ${np.mean(group_a_oil_prices):.2f}")
print(f"Mean Oil Price Group B: ${np.mean(group_b_oil_prices):.2f}")
print(f"Calculated t-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print(f"At α={alpha}, we reject the null hypothesis. There is a significant difference between the mean oil prices of Group A and Group B.")
else:
    print(f"At α={alpha}, we fail to reject the null hypothesis. There is no significant difference between the mean oil prices of Group A and Group B.")
```

#### ANOVA (Analysis of Variance)

ANOVA is used to compare the means of three or more independent groups to determine if at least one group mean is significantly different from the others.

**When to use:** To compare means across three or more independent groups.

**Assumptions:** Independent observations, normal distribution, and homogeneity of variances (variances among groups are equal).

**Example Scenario:** A company wants to compare the sales performance of three different advertising campaigns (Campaign A, B, and C). They run each campaign in different regions and record the average weekly sales. Is there a significant difference in sales performance among the three campaigns?

For demonstration, we'll simulate sales data for three different campaigns.

```python
from scipy import stats
import numpy as np

# Generate synthetic data for three groups
np.random.seed(42)
campaign_a_sales = np.random.normal(loc=100, scale=15, size=30)
campaign_b_sales = np.random.normal(loc=110, scale=15, size=30)
campaign_c_sales = np.random.normal(loc=95, scale=15, size=30)

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(campaign_a_sales, campaign_b_sales, campaign_c_sales)

print(f"Mean Sales Campaign A: {np.mean(campaign_a_sales):.2f}")
print(f"Mean Sales Campaign B: {np.mean(campaign_b_sales):.2f}")
print(f"Mean Sales Campaign C: {np.mean(campaign_c_sales):.2f}")
print(f"Calculated F-statistic: {f_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print(f"At α={alpha}, we reject the null hypothesis. There is a significant difference in sales performance among the campaigns.")
else:
    print(f"At α={alpha}, we fail to reject the null hypothesis. There is no significant difference in sales performance among the campaigns.")
```

#### Chi-Square Goodness-of-Fit (GoF) Test

The Chi-Square Goodness-of-Fit test is used to determine whether a sample distribution matches a population distribution or a theoretical distribution. It's used for categorical data.

**When to use:** To test if observed frequencies of categories differ significantly from expected frequencies.

**Example Scenario:** A geneticist theorizes that a particular gene occurs in a certain population with proportions 25% for allele A, 50% for allele B, and 25% for allele C. A sample of 200 individuals reveals 40 with allele A, 110 with allele B, and 50 with allele C. Does this sample distribution fit the theorized proportions?

For demonstration, we'll create some observed frequencies and a set of expected proportions. We could potentially use a categorical column from our datasets, but they might not have a clear 'expected' distribution to test against, so simulating is clearer.

```python
from scipy import stats
import numpy as np

# Observed frequencies
observed_frequencies = np.array([40, 110, 50]) # Counts for Allele A, B, C

# Total number of observations
total_observations = np.sum(observed_frequencies)

# Expected proportions (from the theory)
expected_proportions = np.array([0.25, 0.50, 0.25])

# Calculate expected frequencies based on total observations and expected proportions
expected_frequencies = expected_proportions * total_observations

# Perform Chi-Square Goodness-of-Fit test
chi2_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")
print(f"Calculated Chi-Square Statistic: {chi2_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print(f"At α={alpha}, we reject the null hypothesis. The observed distribution is significantly different from the expected distribution.")
else:
    print(f"At α={alpha}, we fail to reject the null hypothesis. The observed distribution does not significantly differ from the expected distribution.")
```

#### Chi-Square Independence Test

The Chi-Square Test of Independence is used to determine if there is a significant association (relationship) between two categorical variables. It tests whether the two variables are independent of each other.

**When to use:** To examine the relationship between two categorical variables.

**Example Scenario:** A survey investigates whether there is a relationship between gender (male/female) and preference for a certain political party (Party A/Party B/Other). Do preferences for political parties depend on gender?

For demonstration, we'll create a contingency table (a table of observed frequencies) for two hypothetical categorical variables. We can imagine using columns like `event_type` and `region` from the `geopolitical_df` if they were clean categorical variables, but creating a simpler example is more straightforward for demonstration.

```python
from scipy import stats
import pandas as pd

# Create a contingency table (observed frequencies)
# Example: Relationship between 'Gender' and 'Preferred Political Party'
contingency_table = pd.DataFrame({
    'Party A': [50, 30], # Male, Female
    'Party B': [40, 60], # Male, Female
    'Other': [10, 10]    # Male, Female
}, index=['Male', 'Female'])

print("Contingency Table (Observed Frequencies):")
display(contingency_table)

# Perform Chi-Square Test of Independence
chi2_statistic, p_value, degrees_of_freedom, expected_frequencies = stats.chi2_contingency(contingency_table)

print(f"\nCalculated Chi-Square Statistic: {chi2_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print("Expected Frequencies:")
print(expected_frequencies)

alpha = 0.05
if p_value < alpha:
    print(f"\nAt α={alpha}, we reject the null hypothesis. There is a significant association between Gender and Political Party Preference.")
else:
    print(f"\nAt α={alpha}, we fail to reject the null hypothesis. There is no significant association between Gender and Political Party Preference (they are independent).")
```
